```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
print(data.shape)
data.head()
```

```python
data.describe()
```

```python
num_unique = len(data["cohesion"].value_counts())

nrows = 2
ncols = 3
fig, axes = plt.subplots(nrows, ncols, figsize=(12, 8))
columns = ['cohesion', 'syntax', 'vocabulary', 'phraseology', 'grammar', 'conventions']

for idx, column in enumerate(columns):
    row, col = idx // ncols, idx % ncols
    num_unique = len(data[column].value_counts())
    axes[row, col].hist(data[column], bins=num_unique)
    axes[row, col].set_title(column)
    
    if row == nrows - 1:
        axes[row, col].set_xlabel("Scores from 1-5")

    if col == 0:
        axes[row, col].set_ylabel("Frequency of score")

fig.suptitle("Score Distribution for Various Metrics", fontsize=16)
fig.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust the tight_layout to account for the suptitle
plt.show()
```
